```python
import math

def calc_avg_dist(x, y):
    assert(len(x) == len(y))
    dists = []
    for i in range(1, len(x)):
        d = math.sqrt((x[i]-x[i-1])**2 + (y[i]-y[i-1])**2)
        dists.append(d)
        
    return min(dists), sum(dists)/len(dists), max(dists)
```

```python
print(calc_avg_dist(x,y))
```
